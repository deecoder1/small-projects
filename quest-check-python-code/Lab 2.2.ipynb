## 1. Python enumerate and zip (Built-In Sequence Functions)
Python has a handful of useful sequence functions that you should familiarize yourself with and use at any opportunity.

### 1.1 enumerate

It’s common when iterating over a sequence to want mto keep track of the index of the current item. 

A do-it-yourself approach would look like:



Since this is so common, Python has a built-in function, `enumerate`, which returns a sequence of `(index, value)` tuple.

In the following example, `enumerate(fruits)` creates an iterator that yields pairs of index and value for each element in the `fruits` list. The `for` loop then iterates over these pairs, printing the index and corresponding fruit for each iteration:

```python
# Create a list of fruits
fruits = ['apple', 'banana', 'cherry', 'orange']

# Use enumerate to get the index and value of each fruit
for index, fruit in enumerate(fruits):
    print("{}: {}".format(index, fruit))
```

```python
for fruit in fruits:
    print(fruit)
```

### 1.2 zip
zip “pairs” up the elements of a number of lists, tuples, or other sequences to create a list of tuples:

```python
seq1 = ["foo", "bar", "baz"]
seq2 = ["one", "two", "three"]

zipped = zip(seq1, seq2)
newlist = list(zipped)

print(newlist)  # output = list of tuples
```

```python
newlist
```

A common use of zip is simultaneously iterating over multiple sequences, possibly also combined with enumerate:

```python
# zip is simultaneously iterating over 2 lists
for index, (a, b) in enumerate(zip(seq1, seq2)):
   print("{}: {} {}".format(index, a, b))
```

## 2 Lambda Functions

### 5.8 Boolan Indexing

Consider an example where we have some data in an array and an array of names 
with duplicates:

```python
names = np.array(["Bob", "Joe", "Will", "Bob", "Will", "Joe", "Joe"])
data = np.array([[4, 7], [0, 2], [-5, 6], [0, 0], [1, 2], [-12, -4], [3, 4]])
```

```python
names
```

```python
data
```

Suppose each name corresponds to a row in the data array and we wanted to select all the rows with the corresponding name "Bob". 

Comparing `names` with the string "Bob" yields a Boolean array:

```python
names == "Bob"
```

This Boolean array can be used to index the data array, i.e. to find all elements in `data` that correspond to the name "Bob" in `names`:

```python
data[names == "Bob"]
```

As a final note, the Boolean array must be of the same length as the array axis it’s indexing. 

The  ~  operator can be useful when you want to invert a Boolean array referenced by a
variable:

```python
cond = names == "Bob"

# find data for names NOT equal to "Bob"
data[~cond]
```

### 5.9  Element-Wise Array Functions (Universal Functions ufuncs)

Numpy has many function like `np.sqrt`, `np.abs()`,  `np.log()`, `np.exp`, `np.sin()`, `np.cos()`,  etc.

### 5.10 Mathematical and Statistical Methods (Aggregations Functions)

NumPy also provides some aggregation functions, including `max()`, `min()`, `sum()`, 'std', and `count_nonzero()`. You can apply these functions about an entire array or about the data along an axis.
 




Notethat these functions` can be called in two different ways: either as methods of an array, or as numpy functions.

##### 1-D arrays

Here are examples of aggregation functions used with a 1-D array.

```python
data = np.array([1, 2, 3])
```

```python
# Calling using methods of array
print( data.min() )  
print( data.max() )
print( data.sum() )
```

```python
# Calling as numpy functions (identical outcome).
print( np.min(data) )  
print( np.max(data) )
print( np.sum(data) )  
```

```python
# Counting elements based on condition 
data > 1
```

```python
np.count_nonzero(data > 1)
```

##### 2-D arrays

Here are examples of aggregation functions used with a 2-D array.

```python
data = np.array([[1, 2], [3, 4], [5, 6]])
data
```

```python
print( data.min() )  
print( data.max() )
print( data.sum() )
print( data.mean() )  # calculate the average.
```

The `axis` parameter can also be used here to perform aggregation along a dimension of an array.

```python
data.max(axis=0)
```

### Exercise 5.3 (Indexing and Slicing)

Create a new 2-D array initialised with the data values `[[0, 5, 23], [37, 42, 54], [61, 79, 88]]`.

Use indexing to access the data value in second row, third column of the array.

Use slicing to access the first and second rows of the array.

Access the last two values in the second column of the array.


```python
import numpy as np

# Create the 2D NumPy array
arr = np.array([
    [0, 5, 23], 
    [37, 42, 54], 
    [61, 79, 88]
])

# Access the value in the second row, third column (index [1, 2])
second_row_third_col = arr[1, 2]
print(second_row_third_col)

# Slice the first and second rows (rows 0 and 1)
first_two_rows = arr[:2, :]
print(first_two_rows)

# Access the last two values in the second column (rows 1 and 2, column 1)
last_two_values_second_col = arr[1:, 1]
print(last_two_values_second_col)
```

### Exercise 5.4 (Stats Methods)

Create a 2-D array of random numbers with 4 rows and 3 columns.

Multiply the array by 100.

Find the maximum, minimum and sum of the array.

Find the maximum value in each row.

Find the sum and mean values in each column.


```python
import numpy as np

# Create a 2D array of random numbers (4x3)
arr = np.random.rand(4, 3)  # Generates random numbers between 0 and 1

# Multiply the array by 100
arr *= 100

# Find max, min, and sum of the array
max_value = np.max(arr)
min_value = np.min(arr)
sum_value = np.sum(arr)

# Find the maximum value in each row
max_per_row = np.max(arr, axis=1)

# Find the sum and mean values in each column
sum_per_col = np.sum(arr, axis=0)
mean_per_col = np.mean(arr, axis=0)

# Print results
print(arr)
print(max_value)
print(min_value)
print(sum_value)
print(max_per_row)
print(sum_per_col)
print(mean_per_col)
```
